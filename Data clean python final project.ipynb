```python
import pandas  as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\Dell\Desktop\python final project dataset\spotify dataset.csv")
```

```python
df.head()
```

```python
df.describe
```

```python
df.isnull().sum()
```

```python
df['artists'].fillna("Unknown", inplace=True)
df['album_name'].fillna("Unknown", inplace=True)
df['track_name'].fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.hist(df['popularity'], bins=20, color='red')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Distribution of Popularity')
plt.show()
```

```python
genre_counts = df['track_genre'].value_counts()
plt.bar(genre_counts.index, genre_counts.values, color='coral')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Distribution')
plt.xticks(rotation=90)
plt.show()
```
